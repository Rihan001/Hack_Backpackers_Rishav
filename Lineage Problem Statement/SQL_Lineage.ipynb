```
from sql_metadata import Parser
def get_resultantColumnWithTableName(sql_query):
  parse = Parser(sql_query)
  table_column = parser.columns_dict.get('select')
  if(table_column):
    for i in table_column:
      a = i.split('.')
      print(a[1]+ " => "+a[0])
```

```
get_resultantColumnWithTableName("SELECT o.OrderID, c.CustomerName, o.OrderDate FROM Orders o INNER JOIN Customers c ON o.CustomerID=c.CustomerID")
```
